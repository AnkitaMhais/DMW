```python
import pandas as pd
import seaborn as sns
sns.set(color_codes=True)
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('/content/weatherHistory.csv')
```

```python
#Barplot of humidity and temperature
sns.barplot(data['Humidity'],data['Temperature (C)'])
```

```python
#Distplot for continous data representation
sns.distplot(data['Humidity']) 
```

```python
#Bivariate functions(More than one column)
sns.jointplot(data['Humidity'],data['Temperature (C)'])
```

```python

sns.distplot(data['Humidity'], kde=False, rug=True);
```

```python
#Data visualizing instead of dots with hexagon
sns.jointplot(data['Humidity'], data['Temperature (C)'], kind="hex")
```

```python
#Ploting categorical data
sns.stripplot(data['Humidity'],data['Pressure (millibars)'])
```

```python
sns.boxplot(data['Humidity'], data['Temperature (C)'])
```

```python
sns.barplot(data['Humidity'], data['Temperature'], hue=data['Wind Speed (km/h'])
```

```python
sns.countplot(data['Humidity'])
```

```python
#lmplot used for linear regression
sns.lmplot(x="Humidity", y="Temperature (C)", data=data)
```
